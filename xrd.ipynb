```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("TnDJGCQnzv8uX3Ao") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-985829")
#structure = Structure.from_file("HfS2_symmetrized.cif")
xrdc = XRDCalculator("CuKa1") #Pour le CuKa de longueur d’onde λ=1.54060 Å, le plus proche est CuKa1 de longueur d’onde λ=1.54056 Å
xrdpat = xrdc.get_pattern(structure)
```

```python
print('Premier pic : ')
print('2θ = {}°'.format((xrdpat.x)[0]))
print('Intensité (par rapport au plud haut pic) = {}%'.format((xrdpat.y)[0]))
print("hkl :",xrdpat.hkls[0][0]['hkl'])
print("Multiplicité =",xrdpat.hkls[0][0]['multiplicity'])
print("Distance inter-plan = {}Å".format(xrdpat.d_hkls[0]))
```

```python
print('Deuxième pic : ')
print('2θ = {}°'.format((xrdpat.x)[1]))
print('Intensité (par rapport au plud haut pic) = {}%'.format((xrdpat.y)[1]))
print("hkl :",xrdpat.hkls[1][0]['hkl'])
print("Multiplicité =",xrdpat.hkls[1][0]['multiplicity'])
print("Distance inter-plan = {}Å".format(xrdpat.d_hkls[1]))
```

```python
print('Troisième pic : ')
print('2θ = {}°'.format((xrdpat.x)[2]))
print('Intensité (par rapport au plud haut pic) = {}%'.format((xrdpat.y)[2]))
print("hkl :",xrdpat.hkls[2][0]['hkl'])
print("Multiplicité =",xrdpat.hkls[2][0]['multiplicity'])
print("Distance inter-plan = {}Å".format(xrdpat.d_hkls[2]))
```
