```python
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from IPython.display import Image
```

```python
with MPRester("TnDJGCQnzv8uX3Ao") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-985829")
#structure = Structure.from_file("HfS2_symmetrized.cif")
sga_s = SpacegroupAnalyzer(structure)
sym = sga_s.get_symmetry_operations()
view = quick_view(structure)
display(view)
```

```python
print("Première symétrie : Inversion")
print("Matrice de symétrie : {}".format(sym[1]))
print("Appliquée à l'atome de Hf #4 de coordonnée : (0.66667, 0.33333, 0.78055)")
print("Nous fournit l'atome de coordonnée : {}".format(sym[1].operate((0.66667, 0.33333, 0.78055))))
#view.script("draw symop 2 {atomno=4}")
Image(filename = 'Symmetry1.png')
```

```python
print("Deuxième symétrie : Rotation")
print("Matrice de symétrie : {}".format(sym[2]))
print("Appliquée à l'atome de Hf #2 de coordonnée : (0.33333, 0.66667, 0.21945)")
print("Nous fournit l'atome de coordonnée : {}".format(sym[2].operate((0.33333, 0.66667, 0.21945))))
#view.script("draw symop 3 {atomno=2}")
Image(filename = 'Symmetry2.png')
```

```python
print("Troisième symétrie : Roto-inversion")
print("Matrice de symétrie : {}".format(sym[3]))
print("Appliquée à l'atome de S #2 de coordonnée : (0.0, 0.1, 0.1)")
print("Nous fournit l'atome de coordonnée : {}".format(sym[2].operate((0.0, 0.1, 0.1))))
#view.script("draw symop 4 {atomno=5}")
Image(filename = 'Symmetry3.png')
```
